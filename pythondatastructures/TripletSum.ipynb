```python
def tripletSum(arr,sm):
    arr.sort()
    sz = len(arr)
    for i in range(0, sz):
        l = i + 1
        r = sz-1 
        while (l < r): 
            if arr[i]+arr[l]+arr[r] == sm:
                print(arr[i],arr[l],arr[r])
                l+=1
            elif arr[i]+arr[l]+arr[r] < sm:
                l+=1
            else:
                r-=1
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
sm=int(input())
tripletSum(arr, sm)
```
